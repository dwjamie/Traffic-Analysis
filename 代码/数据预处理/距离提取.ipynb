```python
from math import radians, sin, cos, asin, sqrt
from pyspark.sql import SparkSession, functions as F
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.parquet('../../数据/ticket_cleaned.parquet')
df.show()
df.count()
```

```python
stations = spark.read.parquet('../../数据/stations.parquet')
stations.show()
stations.count()
```

```python
df = df.join(stations, df.乘车站名称 == stations.车站名称)
df = df.drop('车站名称')
df = df.withColumnRenamed('车站省份', '乘车站省份')
df = df.withColumnRenamed('车站行政地理分区', '乘车站行政地理分区')
df = df.withColumnRenamed('车站经度', '乘车站经度')
df = df.withColumnRenamed('车站纬度', '乘车站纬度')

df = df.join(stations, df.到达站名称 == stations.车站名称)
df = df.drop('车站名称')
df = df.withColumnRenamed('车站省份', '到达站省份')
df = df.withColumnRenamed('车站行政地理分区', '到达站行政地理分区')
df = df.withColumnRenamed('车站经度', '到达站经度')
df = df.withColumnRenamed('车站纬度', '到达站纬度')

df.show()
df.write.parquet('../../数据/ticket_preprocessed_full.parquet')
```

```python
df = df[(df.乘车站省份 != 'N') & (df.到达站省份 != 'N') & (df.乘车站行政地理分区 != 'N') & (df.到达站行政地理分区 != 'N')]
df = df[(df.乘车站经度 != -1.0) & (df.到达站经度 != -1.0) & (df.乘车站纬度 != -1.0) & (df.到达站纬度 != -1.0)]

df.count()
```

```python
def get_distance(lon1, lat1, lon2, lat2):
    # 提取距离
    lon1, lat1, lon2, lat2 = map(radians, (lon1, lat1, lon2, lat2))
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return round(2 * 6371 * asin(sqrt(a)), 3)

get_distance = F.udf(get_distance, 'float')
```

```python
df = df.withColumn('距离', get_distance('乘车站经度', '乘车站纬度', '到达站经度', '到达站纬度'))
df.show()
df.write.parquet('../../数据/ticket_preprocessed.parquet')
```
