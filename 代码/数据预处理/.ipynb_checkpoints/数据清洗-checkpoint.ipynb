```python
from pyspark.sql import SparkSession, functions as F
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.csv('../../数据/ticket.csv')
df.show()
df.count()
```

```python
df = df.drop('_c3', '_c4')
df = df.toDF('班次代码', '发车日期', '发车时间', '乘车站名称', '到达站名称', '座位类型')
df.show()
```

```python
spark.conf.set('spark.sql.legacy.timeParserPolicy', 'CORRECTED')
df = df.withColumn('发车时间', F.to_timestamp(F.concat_ws(' ', '发车日期', '发车时间'), 'yyyyMMdd Hmmss'))  # 将发车日期和时间整合为时间戳
df = df.drop('发车日期')
df.show()
```

```python
df = df[df.座位类型.isin('1', '2', '3', '4', '9')]
df.count()
```

```python
df = df.dropna()
df.count()
```

```python
df.show(1000)
```

```python
df.write.parquet('../../数据/ticket_cleaned.parquet')
```

```python
spark.stop()
```
