```python
from pyspark.sql import SparkSession, functions as F
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.parquet('../../数据/ticket_preprocessed.parquet')
df.show()
df.count()
```

```python
# 年份
df.select(F.year('发车时间')).show()
```

```python
# 月份
df.select(F.month('发车时间')).show()
```

```python
# 日期
df.select(F.dayofmonth('发车时间')).show()
```

```python
# 星期几
df.select(F.dayofweek('发车时间')).show()
```

```python
# 小时
df.select(F.hour('发车时间')).show()
```

```python
spark.stop()
```
